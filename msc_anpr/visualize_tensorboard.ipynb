```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


objec_detection_train = './output/object_detection_logs_train.csv'
objec_detection_validation = './output/object_detection_logs_validation.csv'
objec_detection_validation_iterations = './output/object_detection_logs_validation_iterations.csv'

# Read CSV
df = pd.read_csv(objec_detection_validation_iterations, delimiter=',')

# Set seaborn style
sns.set_style("whitegrid")

# Plot
plt.figure(figsize=(10,6))
sns.lineplot(x='Step', y='Value', data=df, linewidth=2.5)

# Set labels and title
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.title('Evaluation loss over iterations' , fontsize=16)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

objec_detection_train = './output/object_detection_logs_train.csv'
objec_detection_validation = './output/object_detection_logs_validation.csv'

# Load training loss data
df_train = pd.read_csv(objec_detection_train, delimiter=',')
# Load validation loss data over epochs
df_val = pd.read_csv(objec_detection_validation, delimiter=',')

plt.figure(figsize=(10, 6))

# Plot training loss
plt.plot(df_train['Step'], df_train['Value'], label='Training Loss')

# Plot validation loss
plt.plot(df_val['Step'], df_val['Value'], label='Validation Loss')

plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

objec_detection_train = './output/object_detection_logs_train.csv'
objec_detection_validation = './output/object_detection_logs_validation.csv'

# Load the training loss data
df_train = pd.read_csv(objec_detection_train, delimiter=',')

# Load the validation loss data
df_val = pd.read_csv(objec_detection_validation, delimiter=',')

# Create a trace for the training data
trace_train = go.Scatter(
    x = df_train['Step'],
    y = df_train['Value'],
    mode = 'lines',
    name = 'Training loss'
)

# Create a trace for the validation data
trace_val = go.Scatter(
    x = df_val['Step'],
    y = df_val['Value'],
    mode = 'lines',
    name = 'Validation loss'
)

# Create a layout
layout = go.Layout(
    title = 'Training and Validation Loss Over Epochs',
    xaxis = dict(title = 'Epoch'),
    yaxis = dict(title = 'Loss'),
)

# Combine the traces in the data list
data = [trace_train, trace_val]

# Create the figure with the data and layout
fig = go.Figure(data=data, layout=layout)

# Show the figure
fig.show()
```

```python
import pandas as pd
import plotly.express as px

objec_detection_train = './output/object_detection_logs_train.csv'
objec_detection_validation = './output/object_detection_logs_validation.csv'

# Read CSV
df = pd.read_csv(objec_detection_train, delimiter=',')

# Create a line plot
fig = px.line(df, x='Step', y='Value', title='Training Epoch Loss')

# Update layout
fig.update_layout(
    autosize=False,
    width=800,
    height=600,
    xaxis_title="Epoch",
    yaxis_title="Loss",
)

# Show the plot
fig.show()

# Read CSV
df = pd.read_csv(objec_detection_validation, delimiter=',')

# Create a line plot
fig = px.line(df, x='Step', y='Value', title='Validation Epoch Loss')

# Show the plot
fig.show()
```
